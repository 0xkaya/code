# Environment

The CartPole-v1 task involves balancing a pole on a cart. The agent needs to learn to keep the pole upright by applying forces (left or right) to the cart. Gymnasium provides the environment API to interact with this task.

# Q-Learning Implementation

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

# Initialize the CartPole-v1 environment
env = gym.make("CartPole-v1")

# Q-Learning Parameters
state_bins = [20, 20, 20, 20]  # Discretize each state variable into bins
q_table = np.random.uniform(low=-2, high=2, size=(state_bins + [env.action_space.n]))
learning_rate = 0.1
discount_factor = 0.99
epsilon = 1.0  # Exploration rate
epsilon_decay = 0.995
min_epsilon = 0.01
episodes = 5000

# Discretize the state space
def discretize_state(state, bins):
    state_lower_bounds = env.observation_space.low
    state_upper_bounds = env.observation_space.high
    state_upper_bounds[1] = 5  # Clip velocity
    state_upper_bounds[3] = 5  # Clip angular velocity

    state_bins_range = [
        np.linspace(state_lower_bounds[i], state_upper_bounds[i], bins[i] - 1)
        for i in range(len(state))
    ]
    state_indices = [
        np.digitize(state[i], state_bins_range[i]) for i in range(len(state))
    ]
    return tuple(state_indices)

# Train the Q-learning agent
rewards = []
for episode in range(episodes):
    state, _ = env.reset()
    state = discretize_state(state, state_bins)
    total_reward = 0
    done = False

    while not done:
        # Epsilon-greedy action selection
        if np.random.random() < epsilon:
            action = env.action_space.sample()  # Explore
        else:
            action = np.argmax(q_table[state])  # Exploit

        # Take the action
        next_state, reward, done, truncated, _ = env.step(action)
        next_state = discretize_state(next_state, state_bins)
        total_reward += reward

        # Update Q-value
        if not done:
            best_next_action = np.argmax(q_table[next_state])
            q_table[state + (action,)] += learning_rate * (
                reward + discount_factor * q_table[next_state + (best_next_action,)]
                - q_table[state + (action,)]
            )
        else:
            q_table[state + (action,)] += learning_rate * (reward - q_table[state + (action,)])

        state = next_state

    # Update exploration rate
    epsilon = max(min_epsilon, epsilon * epsilon_decay)
    rewards.append(total_reward)

    # Print progress
    if (episode + 1) % 500 == 0:
        print(f"Episode: {episode + 1}, Average Reward: {np.mean(rewards[-500:])}")

env.close()
```

# Visualizing Training Performance

```python
# Plot the rewards over episodes
plt.figure(figsize=(12, 6))
plt.plot(rewards, label="Total Reward")
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.title("Training Performance of Q-Learning on CartPole")
plt.grid()
plt.legend()
plt.show()
```

# Testing the Trained Agent

```python

# Test the trained agent
state, _ = env.reset()
state = discretize_state(state, state_bins)
done = False
total_reward = 0

print("Testing the agent...")
while not done:
    env.render()  # Visualize the environment
    action = np.argmax(q_table[state])  # Always exploit during testing
    next_state, reward, done, truncated, _ = env.step(action)
    state = discretize_state(next_state, state_bins)
    total_reward += reward

print("Total reward in test episode:", total_reward)
env.close()
```

# Slow Motion Visualization During Testing

```python
import gymnasium as gym
import time

# Create the environment with rendering
env = gym.make('CartPole-v1', render_mode="human")

# Test the trained agent
state, _ = env.reset()
state = discretize_state(state, state_bins)
done = False
total_reward = 0

print("Testing the agent...")
while not done:
    env.render()  # Visualize the environment
    time.sleep(0.05)  # Slow down the visualization
    action = np.argmax(q_table[state])  # Always exploit during testing
    next_state, reward, done, truncated, _ = env.step(action)
    state = discretize_state(next_state, state_bins)
    total_reward += reward

print("Total reward in test episode:", total_reward)
env.close()
```

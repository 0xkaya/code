# Autoencoder Example Using MNIST

## Importing Packages

```python
import tensorflow as tf 
from tensorflow.keras.layers import Input, Dense, Flatten, Reshape
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt 
import numpy as np
```

# Data Preperation

```python
(X_train,_),(X_test,_) = tf.keras.datasets.mnist.load_data()
```

```python
# Normalize the data to range [0,1]
X_train = X_train.astype("float32")/255.0
X_test = X_test.astype("float32")/255.0
```

```python
# Flatten the 28X28 images into vectors of size 784
X_train = X_train.reshape(-1,28*28)
X_test = X_test.reshape(-1,28*28)
```

```python
# Autoencoder dimensions
input_dim = X_train.shape[1] # 784
encoding_dim = 64 # Compressed representation dimension
```

```python
## Encoder
```

```python
input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
```

```python
## Decoder
```

```python
decoded = Dense(input_dim, activation='sigmoid')(encoded)
```

```python
# Autoencoder Model
```

```python
autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
## Train the Autoencoder
```

```python
history = autoencoder.fit(
    X_train,
    X_train,
    epochs = 20,
    batch_size =256,
    shuffle = True,
    validation_data =(X_test,X_test)
)
```

```python
# Encoder Model (for compressed presentation)
encoder = Model(inputs=input_layer, outputs=encoded)
```

```python
# Decode Model (for reconstruction)
encoded_input = Input(shape=(encoding_dim,))
decoder_layer = autoencoder.layers[-1]
decoder = Model(inputs=encoded_input,outputs=decoder_layer(encoded_input))
```

```python
# Encode and Decode the Test Data
encoded_imgs = encoder.predict(X_test)
decoded_imgs = decoder.predict(encoded_imgs)
```

```python
# Visualization
```

```python
## Plot Training Loss
plt.figure(figsize=(8,4))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label ="Validation Loss")
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
## Visualize Original, Encoded, and Reconstructed Images
```

```python
n = 10 # Number of images to display
plt.figure(figsize=(18,6))

for i in range(n):
    # Original images
    plt.subplot(3,n,i+1)
    plt.imshow(X_test[i].reshape(28,28), cmap="gray")
    plt.axis("off")
    if i == n//2:
        plt.title("Original Images")
    
    # Encoded Representations
    plt.subplot(3,n,i+n+1)
    plt.imshow(encoded_imgs[i].reshape(8,8), cmap="viridis")
    plt.axis("off")
    if i == n//2:
        plt.title("Encoded Images")
        
    # Reconstructed images
    plt.subplot(3, n, i + 2 * n + 1)
    plt.imshow(decoded_imgs[i].reshape(28, 28), cmap="gray")
    plt.axis("off")
    if i == n // 2:
        plt.title("Reconstructed Images")
        
plt.tight_layout()
plt.show()
    
```
